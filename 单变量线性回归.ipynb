# 单变量线性回归算法 x 代表学历， f(x) 代表收入
# f(x) = ax + b
# 我们使用 f(x) 这个函数来映射输入特征和输出值

# 预测目标与损失函数

```python
x = data.Education
y = data.Income
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_shape=(1,)))
```

```python
# ax + b
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
# 模型训练
history = model.fit(x, y, epochs=5000)
```

```python
# 模型预测
model.predict(x)
```
